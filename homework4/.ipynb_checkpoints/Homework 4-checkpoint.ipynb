```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate


def create_gram(N):
    phi = lambda z: np.exp(-z**2)
    gram_matrix = np.zeros((N,N))    
    
    for i in range(N):
        for j in range(N):
            matrix_elem = lambda z: phi(N*z - i + 0.5)*phi(N*z - j + 0.5)
            gram_matrix[i][j] = integrate.quad(matrix_elem, -np.inf, np.inf)[0]
            
    return gram_matrix
            
def plot_basis(N, n):
    G = create_gram(N)
    H = np.linalg.inv(G)
    
    x = np.arange(0,1,0.001)
    y = np.zeros(len(x))
    
    for t in range(len(y)):
        y[t] = dual_basis(N, n, x[t], G)
    
    plt.plot(x, y)
    plt.show()
    
def dual_basis(N, n, t, H):
    phi = lambda x: np.exp(-x**2)
    
    dual_basis = 0
    
    for l in range(N):
        temp_phi = phi((N*t) -(l+1) + 0.5)
        #print("temp_phi t={} l={} = {}".format(t, l, temp_phi))
        dual_basis += (H[n][l])*temp_phi
    
    return dual_basis

def plot_kernel(N, tau):
    phi = lambda x: np.exp(-x**2)
    G = create_gram(N)
    H = np.linalg.inv(G)
    
    x = np.arange(0,1,0.001)
    y = np.zeros(len(x))
    
    for t in range(len(y)):
        for n in range(N):
            y[t] += phi()*dual_basis(N, n, x[t], G)
    
    plt.plot(x, y)
    plt.show()
    
```

```python
plot_basis(32, 18)
```

```python
import scipy.io

mat = scipy.io.loadmat('hw4p4.mat')

u_data = mat['udata']

y_data = mat['ydata']
```

```python
def generate_A_matrix(u_data):
    basis = [lambda s,t: s**2, lambda s,t: t**2, lambda s,t: s*t, lambda s,t: s, lambda s,t: t, lambda s,t: 1]
    
    
    A = np.zeros((100,6))
    
    for i in range(100):
        for j in range(6):
            A[i][j] = basis[j](u_data[0][i], u_data[1][i])
            
    return A
```

```python
def solve_least_squares(u_data, y_data):
    A = generate_A_matrix(u_data)
    A_t = np.transpose(A)
    
    alphas = np.matmul(A_t, y_data)
    
    return alphas
    
alphas = solve_least_squares(u_data, y_data)
```

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

def contour_plot(alphas):
    fig = plt.figure()
    ax = fig.gca(projection='3d')
    
    X = np.arange(0,1,0.02)
    Y = np.arange(0,1,0.02)
    
    X, Y = np.meshgrid(X, Y)
    
    Z = alphas[0]*X**2 + alphas[1]*Y**2 + alphas[2]*X*Y + alphas[3]*X + alphas[4]*Y + alphas[5]
    
    surface = ax.plot_surface(X, Y, Z)
        
    plt.show()
    
contour_plot(alphas)
```
